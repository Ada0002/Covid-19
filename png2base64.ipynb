```python
import base64
import os.path
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
for i in range (100):
  # Reading from file
  with open(f"/content/drive/MyDrive/Covid 19/Newron Covid 19/Covid/COVID-{i+1}.png", "rb+") as imageFile:
    encodedString = base64.b64encode(imageFile.read())
  
  print(encodedString);

  # Write Image from Base64 File
  save_path = '/content/Covid 19/Covid'
  completeName = os.path.join(save_path, f"COVID-{i+1}"".txt")  
# Writing to file
  with open(completeName, "wb") as file1:
    # Writing data to a file
    file1.write(encodedString)
```
